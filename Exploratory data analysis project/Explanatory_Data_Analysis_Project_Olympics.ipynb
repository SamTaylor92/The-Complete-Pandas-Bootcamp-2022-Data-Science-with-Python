## [Data cleaning] Merging and concatenating

1. __Merge__ summer and winter (one row for each medal awarded in any Olympic games) and save the merged dataframe in __olympics__.

    1. __summer__ or __winter__.

```python
# Append the summer and winter datasets together
## Create a column (Edition) which tells us what dataset the data is from
## Reset the index and drop the original index
pd.concat([summer, winter], axis = 0, keys = ["Summer", "Winter"],
          names = ["Edition"]).reset_index().drop(columns = "level_1")
```

```python
# Save the dataframe to a variable
olympics = pd.concat([summer, winter], axis = 0, keys = ["Summer", "Winter"],
                     names = ["Edition"]).reset_index().drop(columns = "level_1")
```

## [Analysis] Hypothesis testing with scipy

Work with __rank()__ to see if the following variables have an effect on __total medals__:
- __GDP__
- __Total_Games__ 
- __Population__

__Note:__ As the data is ranked, we will be using a Spearman's correlation 

```python
# Inspect
medals_per_country
```

```python
# Drop the individual medal columns
medals_per_country.drop(columns = ["Gold", "Silver", "Bronze"], inplace = True)
```

```python
# Correlation analysis with 'pearson'
medals_per_country.corr(method = "pearson")
```

```python
# Correlation analysis with 'Spearman'
medals_per_country.corr(method = "spearman")
```

```python
# Correlation analysis with 'pearson'
medals_per_country.rank(ascending = False).corr(method = "pearson")
```

```python
# Import library
import scipy.stats as stats
```

------------------------

### [Hypothesis 1] There is no relationship between `Total Medals` and `Population`

```python
##
stats.spearmanr(medals_per_country.Total, medals_per_country.Population, 
                             nan_policy = "omit")
```

```python
# Assign variables r to correlation coefficiant & p_value to pvalue
r, p_value = stats.spearmanr(medals_per_country.Total, medals_per_country.Population, 
                             nan_policy = "omit")
```

```python
round(p_value)
```

#### ❌ Reject Hypothesis 1: 
There is a significant (positive) relationship between `Total Medals` and `Population`

--------------------------------

```python
r, p_value = stats.spearmanr(medals_per_country.Total, medals_per_country.GDP,
                             nan_policy = "omit")
```

```python
p_value
```

#### ❌ Reject Hypothesis 2:

## [Analysis] Aggregating and Ranking

Create the following __Seaborn Heatmap__ with __Medal Ranks__ for Top 50 Countries: 
- Total medals
- Summer games medals
- Winter games medals 
- Men 
- Women

```python
# Insepect dataset
olympics
```

```python
# Pivot dataframe (country vs summer/winter)
sum_vs_win = pd.crosstab(olympics.Country, olympics.Edition)
```

```python
# Inspect
sum_vs_win
```

```python
# Pivot dataframe (country vs gender)
men_vs_women =  pd.crosstab(olympics.Country, olympics.Gender)

# Inspect
men_vs_women
```

```python
# Join newly created dataframes together (column concat)
medals_by_cats = pd.concat([sum_vs_win, men_vs_women], axis = 1)

# Inspect
medals_by_cats 
```

```python
# Add total column
medals_by_cats["Total"] = medals_by_cats.Summer + medals_by_cats.Winter
```

```python
# Inspect
medals_by_cats
```

```python
# Sort values by total column
medals_by_cats.sort_values("Total", ascending = False, inplace = True)
```

```python
# Inspect
medals_by_cats
```

```python
# Rank the dataframe (each column ranked separately)
ranks = medals_by_cats.rank(ascending = False, method = "average")
ranks
```

```python
# Slice dataframe for top 50
top_50 = ranks.head(50)
```

```python
# Create a heatmap

# Set chart size
plt.figure(figsize = (50, 10))

# Create heatmap (select dataframe, select colour coding, add annotation, set format of numbers)
sns.heatmap(top_50.T,cmap='RdYlGn_r',annot=True,fmt='2.0f')

# Display chart
plt.show()
```
